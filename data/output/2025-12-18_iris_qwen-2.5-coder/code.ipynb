```
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])

# Display the first few rows of the DataFrame
print(df.head())

# Summary statistics for each column
print(df.describe())

# Check the distribution of target variable
print(df['target'].value_counts())

# Correlation matrix to understand relationships between features
correlation_matrix = df.corr()
print(correlation_matrix)

# Pairplot to visualize pairwise relationships between features and target
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, hue='target')
plt.show()

# Box plot for each feature to visualize the distribution of data points
df.boxplot(by='target', layout=(2, 2))
plt.show()
```

```
import pandas as pd

# Load the Iris dataset
iris = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')

# Display the first few rows of the dataset
print(iris.head())

# Get summary statistics for each column
print(iris.describe())

# Select numerical columns and compute correlation matrix
numerical_cols = iris.select_dtypes(include=[np.number]).columns
correlation_matrix = iris[numerical_cols].corr()
print(correlation_matrix)

# Visualize the correlation matrix using a heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Iris Dataset')
plt.show()

# Plot histograms for each numerical column
iris[numerical_cols].hist(bins=20, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Columns in Iris Dataset', size=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Display the first few rows of the dataset
print(iris.head())

# Summary statistics of the dataset
print(iris.describe())

# Check for missing values
print(iris.isnull().sum())

# Distinguish column types and perform analysis
numerical_cols = iris.select_dtypes(include=[np.number])
categorical_cols = iris.select_dtypes(include=['object'])

# Correlation matrix for numerical columns
correlation_matrix = numerical_cols.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()

# Box plots for numerical columns
numerical_cols.boxplot(figsize=(12, 8))
plt.show()

# Count plot for categorical columns
for column in categorical_cols.columns:
    sns.countplot(x=column, data=iris)
    plt.title(f'Count of {column}')
    plt.show()
```

```
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])

# Display the first few rows of the DataFrame
print(df.head())

# Get summary statistics for numerical columns
print(df.describe(include=[np.number]))

# Get summary statistics for categorical columns
print(df.describe(include=[np.object]))

# Check for missing values
print(df.isnull().sum())

# Correlation matrix for numerical columns
correlation_matrix = df.corr()
print(correlation_matrix)

# Pairplot to visualize relationships between variables
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, hue='target')
plt.show()

# Boxplots to visualize the distribution of each feature by class
for column in iris['feature_names']:
    sns.boxplot(x='target', y=column, data=df)
    plt.title(f'Boxplot of {column} by Class')
    plt.show()
```

```
import pandas as pd

# Load the Iris dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
iris_df = pd.read_csv(url, names=column_names)

# Display summary statistics
print(iris_df.describe())

# Check for missing values
print(iris_df.isnull().sum())
```

```
import seaborn as sns
import matplotlib.pyplot as plt

# Create a pairplot to visualize relationships between features
sns.pairplot(iris_df, hue="class")
plt.show()

# Create boxplots for each feature by class
for column in iris_df.select_dtypes(include=[np.number]).columns:
    sns.boxplot(x="class", y=column, data=iris_df)
    plt.title(f"Boxplot of {column} by Class")
    plt.show()
```
